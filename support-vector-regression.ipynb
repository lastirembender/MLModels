```python
data = pd.read_csv(os.path.join(dirname, "maaslar.csv"))
```

```python
data.head()
```

```python
x = data.iloc[:, 1:2]
y = data.iloc[:, 2:3]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x.values, y.values)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X = sc.fit_transform(x)

sc1 = StandardScaler()
Y = sc1.fit_transform(y)
```

```python
from sklearn.svm import SVR
```

```python
rbf_svr = SVR(kernel='rbf')
```

```python
rbf_svr.fit(X, Y.ravel())
```

```python
plt.scatter(X, Y, color="red")
plt.plot(X, rbf_svr.predict(X), color="blue")
```

```python
lnr_svr = SVR(kernel="linear")
lnr_svr.fit(X, Y.ravel())
```

```python
plt.scatter(X, Y, color="green")
plt.plot(X, lnr_svr.predict(X), color="purple")
```

```python
poly_svr = SVR(kernel="poly")
poly_svr.fit(X, Y.ravel())
```

```python
plt.scatter(X, Y, color="black")
plt.plot(X, poly_svr.predict(X), color="pink")
```

```python
sigmoid_svr = SVR(kernel="sigmoid")
sigmoid_svr.fit(X, Y.ravel())
```

```python
plt.scatter(X, Y, color="green")
plt.plot(X, poly_svr.predict(X), color="brown")
```
